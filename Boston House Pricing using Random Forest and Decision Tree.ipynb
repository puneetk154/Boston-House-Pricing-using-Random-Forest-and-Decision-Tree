## Regression

```python
# Perform the Train and Test Split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Decision Tree for Regression

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
predict = dt.predict(x_test)
```

```python
predict
```

```python
# r2 score for Decision Tree

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, r2_score
print('r2 score:', r2_score(y_test, predict))
```

```python
# Decision Tree for Regression

from sklearn import tree
```

```python
plt.figure(figsize = (80,80))
tree.plot_tree(dt, filled = True)
plt.show()
```

```python
# Random Forest for Regression

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
predict = rf.predict(x_test)
```

```python
predict
```

```python
# r2 score for Random Forest

print('r2 score:', r2_score(y_test, predict))
```

```python
# Boosting - Adaboost for Regression

from sklearn.ensemble import AdaBoostRegressor
ad = AdaBoostRegressor()
ad.fit(x_train, y_train)
```

```python
predict = ad.predict(x_test)
```

```python
predict
```

```python
# r2 score of Adaboost for Regression

print('r2 score:', r2_score(y_test, predict))
```

## Classification

```python
# Select the 'CHAS' column for Classification

data['CHAS'].value_counts()
```

```python
data.head()
```

```python
# Segregate the Dataset 'data' into Independent and Target/Dependent variables

x = data.drop('CHAS', axis=1)
y = data['CHAS']
```

```python
# Check the Independent Variables

x
```

```python
# Check the Target/Dependent Variable

y
```

```python
# Perform the Train and Test Split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# Decision Tree on Classification

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
predict = dt.predict(x_test)
```

```python
predict
```

```python
# Accuracy Score of Decision Tree for Classification

print('Accuracy Score:', accuracy_score(y_test, predict))
```

```python
# Decision Tree for Classification

plt.figure(figsize = (90,90))
tree.plot_tree(dt, filled = True)
plt.show()
```

```python
# Confusion Matrix for Decision Tree - Classification

confusion_matrix(y_test, predict)
```

```python
# Classification Report for Decision Tree - Classification

print(classification_report(y_test, predict))
```
